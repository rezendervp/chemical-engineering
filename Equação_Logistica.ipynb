```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
r_values = np.linspace(1, 6.0, 2000)  # Valores de r (taxa de crescimento)
iterations = 1000  # Nmero de iteraes
last_iterations = 100  # Nmero de iteraes a serem exibidas

# Funo da equao logstica
def logistic_map(r, x):
    return r * x * (1 - x/K)

# Inicializao de arrays para armazenar os resultados
x_values = []
K = 0.5
# Loop atravs dos valores de r
for r in r_values:
    x = 0.1  # Condio inicial
    for _ in range(iterations):
        x = logistic_map(r, x)
        if iterations - last_iterations <= _:
            x_values.append([r, x])

# Criando o grfico de bifurcao
x_values = np.array(x_values)
plt.figure(figsize=(10, 6))
plt.plot(x_values[:, 0], x_values[:, 1], ',k', alpha=0.5)
plt.xlabel('Taxa de Crescimento (r)')
plt.ylabel('Populao')
plt.title('Grfico de Bifurcao da Equao Logstica')
plt.show()
```
