```python
from scipy.integrate import odeint
import numpy as np

# Definir as equações diferenciais
def equations(y, t):
    k = 0.01
    kinv = 0.007
    y1, y2, y3  = y
    dydt = [-k*y1*y2, -k*y1*y2, k*y1*y2 - kinv*y3*y1]
    return dydt

# Condições iniciais
y0 = [1.0, 2, 0]

# Tempo inicial e final
tspan = np.linspace(0, 600, 601)

# Resolver as EDOs usando odeint
sol = odeint(equations, y0, tspan,  atol=1e-10, rtol=1e-8)

# Plotar a solução
import matplotlib.pyplot as plt
plt.plot(tspan, sol[:, 0], 'b', label='y1')
plt.plot(tspan, sol[:, 1], 'r', label='y2')
plt.plot(tspan, sol[:, 2], 'g', label='y3')
plt.legend(loc='best')
plt.xlabel('t[s]')
plt.ylabel('Concentração')
plt.title('Cinética de Reação')
plt.grid()
plt.show()
```

```python
from scipy.integrate import solve_ivp
import numpy as np

# Definir as equações diferenciais
def equations(t, y):
    y1, y2, y3  = y
    dydt = [-y1*y2, -y1*y2, y1*y2-y3**(0.2)]
    return dydt

# Condições iniciais
y0 = [1.2, 1, 0]

# Tempo inicial e final
tspan = [0, 30]

# Resolver as EDOs usando solve_ivp
sol = solve_ivp(equations, tspan, y0,  max_step=0.01)

# Plotar a solução
import matplotlib.pyplot as plt
plt.plot(sol.t, sol.y[0], 'b', label='y1')
plt.plot(sol.t, sol.y[1], 'r', label='y2')
plt.plot(sol.t, sol.y[2], 'g', label='y3')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
